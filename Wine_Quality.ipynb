1.	Duomenų rinkinio aprašymas
Vyno kokybės duomenų rinkinio, kurį sudaro įvairios fizikinės ir cheminės vyno savybės, pvz., rūgštingumas, cukraus kiekis, sieros dioksido lygis ir alkoholio kiekis, analizė. Duomenų rinkinyje yra 1 143 raudonojo vyno pavyzdžiai, kurių kiekvienas pasižymi 12 savybių, įskaitant fiksuotą rūgštingumą, lakiąjį rūgštingumą, citrinų rūgštį, likutinį cukrų, chloridus, laisvą ir bendrą sieros dioksidą, tankį, pH, sulfatus ir alkoholio lygį. Tikslinis kintamasis yra „kokybė“, kiekvienam vyno pavyzdžiui priskiriamas balas nuo 0 iki 10. Pagrindinis šios analizės tikslas yra ištirti ryšį tarp vynų cheminių savybių ir jų kokybės įvertinimų, patikrinant hipotezę, kad šios savybės gali veiksmingai numatyti kokybę. Atliksime išankstinį duomenų apdorojimą, tiriamąją duomenų analizę (įskaitant vizualizacijas, tokias kaip histogramos ir sklaidos diagramos, ir koreliacijos analizę), po to kursime ir įvertinsime nuspėjamuosius modelius.


Štai kiekvieno vyno kokybės duomenų rinkinio stulpelio paaiškinimas:

Fiksuotas rūgštingumas: matuoja nelakiųjų rūgščių (pvz., vyno rūgšties) koncentraciją vyne. Didelis fiksuoto rūgštingumo lygis prisideda prie vyno aitrumo.

Lakioji rūgštingumas: Nurodo acto rūgšties kiekį vyne. Didelis kiekis gali sukelti nemalonų į actą panašų skonį.

Citrinų rūgštis: randama nedideliais kiekiais, citrinos rūgštis gali suteikti vynams „gaivumo“ ir skonio.

Likutis cukrus: tai cukraus kiekis, likęs pasibaigus fermentacijai. Paprastai tai nedidelis kiekis, bet saldesniuose vynuose gali būti didesnis.

Chloridai: druskos kiekis vyne. Didesnis chloridų kiekis gali prisidėti prie sūraus skonio.

Laisvasis sieros dioksidas (SO2): Sieros dioksido dalis, nesusijusi su kitomis molekulėmis ir laisva vyne. Sieros dioksidas veikia kaip antimikrobinis ir antioksidantas.

Bendras sieros dioksido kiekis: bendras sieros dioksido kiekis vyne, įskaitant laisvą ir surištą formą. Jis naudojamas kaip konservantas, apsaugantis nuo gedimo ir oksidacijos.

Tankis: vyno tankis, leidžiantis suprasti jo alkoholio kiekį. Paprastai didesnio alkoholio vynai turi mažesnį tankį.

pH: matuoja vyno rūgštingumą arba šarmiškumą skalėje nuo 0 (labai rūgštus) iki 14 (labai šarminis); daugumos vynų pH skalėje yra 3-4.

Sulfatai: junginys, prisidedantis prie sieros dioksido (SO2) kiekio vyne, veikiantis kaip antimikrobinis ir antioksidantas.

Alkoholis: alkoholio procentas vyne. Didesnis alkoholio kiekis gali turėti įtakos vyno kūnui ir bendram skonio profiliui.

Kokybė: tai tikslinis kintamasis, nurodantis vyno kokybę. Tai balas nuo 0 (prasta kokybė) iki 10 (puikia kokybė), kurį paprastai skiria vyno ekspertai, remdamiesi jutimo testais.

Id: kiekvienos duomenų rinkinio eilutės identifikatorius. Tai paprastai naudojama duomenų tvarkymui ir neturi analitinės reikšmės.

# **Darbo tikslas ir uždaviniai**

Vyno kokybės duomenų rinkinio analizės tikslas yra sukurti nuspėjamuosius modelius, kurie gali tiksliai nustatyti vyno kokybę pagal jo fizikines ir chemines savybes. Užduotys, susijusios su šio tikslo pasiekimu, apima:

•	Išankstinis duomenų apdorojimas: duomenų rinkinio valymas siekiant užtikrinti duomenų kokybę, įskaitant trūkstamų verčių tvarkymą, nukrypimų pašalinimą ir duomenų standartizavimą arba normalizavimą, jei reikia.

•	Tiriamoji duomenų analizė (EDA): pradinis duomenų tyrimas naudojant įvairius statistinius ir vaizdinius metodus, siekiant suprasti pagrindinius modelius, kintamųjų pasiskirstymą ir ryšius tarp skirtingų savybių ir tikslinio kintamojo (vyno kokybės).

•	Funkcijų inžinerija: svarbiausių funkcijų nustatymas ir galbūt naujų funkcijų kūrimas, siekiant pagerinti modelių nuspėjamąją galią.

•	Modelio kūrimas: įvairių mašininio mokymosi modelių, tokių kaip regresijos modeliai ar sudėtingesni neuroniniai tinklai, kūrimas, siekiant numatyti vyno kokybę.

•	Modelio įvertinimas ir parinkimas: kiekvieno modelio našumo įvertinimas naudojant atitinkamą vertinimo metriką, modelio parametrų patikslinimas ir geriausio modelio parinkimas pagal jo tikslumą ir apibendrinimą.

•	Aiškinimas ir ataskaitų teikimas: rezultatų analizavimas, siekiant gauti prasmingų įžvalgų, ypač suprasti, kurios fizikinės ir cheminės savybės daro didžiausią įtaką vyno kokybei.

```python
from sklearn.preprocessing import PolynomialFeatures

# Ensure to work on a copy of the DataFrame to avoid SettingWithCopyWarning
wine_data_mod = wine_data.copy()

# Creating interaction term
wine_data_mod['acid_sugar_interaction'] = wine_data_mod['fixed acidity'] * wine_data_mod['residual sugar']

# Using PolynomialFeatures for squared terms
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(wine_data_mod[['fixed acidity', 'volatile acidity', 'citric acid']])

# Generate feature names for the polynomial features
# Check if 'get_feature_names' is available
if hasattr(poly, 'get_feature_names'):
    new_feature_names = poly.get_feature_names(['fixed acidity', 'volatile acidity', 'citric acid'])
else:
    # If not, create feature names manually
    new_feature_names = ['feature_' + str(i) for i in range(poly_features.shape[1])]

# Add new features to the DataFrame
wine_data_mod = wine_data_mod.join(pd.DataFrame(poly_features, columns=new_feature_names, index=wine_data_mod.index))

# Now you can view the modified DataFrame
wine_data_mod.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.regularizers import l1_l2
from tensorflow.keras.utils import to_categorical

# Sureguliuojama class labels, kad pradėti nuo 0, o ne nuo 3
y_train_adjusted = y_train_resampled - 3
y_test_adjusted = y_test - 3


def create_complex_mlp_model(input_dim, regularization_rate=0.01):
    model = Sequential()
    # Tankaus sluoksnio priejimas su L1 ir L2 reguliavimu
    model.add(Dense(64, input_dim=input_dim, activation='relu',
                    kernel_regularizer=l1_l2(l1=regularization_rate, l2=regularization_rate)))
    # Batch normalizavimo sluoksnis 
    model.add(BatchNormalization())
    # Kitas tankus sluoksnis su dropout
    model.add(Dense(32, activation='relu'))
    model.add(Dropout(0.5))
    # Išvesties sluoksnis su „softmax“ aktyvavimu kelių klasių klasifikavimui
    model.add(Dense(6, activation='softmax'))  # 6 neuronai  6 klasems
    # Sudaromas modelis
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
# Darant prielaidą, kad jūsų duomenys jau yra pakeisti ir suskirstyti į X_train, X_test, y_train, y_test
# Vienkartinis užkoduotas tikslinis kintamasis
# Dabar vienkartinis kodavimas su pakoreguotomis klasių etiketėmis

y_train_encoded = to_categorical(y_train_adjusted, num_classes=6)
y_test_encoded = to_categorical(y_test_adjusted, num_classes=6)
# Sukuriamas modelis
complex_mlp_model = create_complex_mlp_model(input_dim=X_train_resampled.shape[1])

# Pritaikoma modeliui 
complex_mlp_model.fit(X_train_resampled, y_train_encoded, validation_data=(X_test, y_test_encoded), epochs=100, batch_size=32)
```
